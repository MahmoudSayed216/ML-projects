```python
import pre_process as p_p

import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
from math import sqrt
```

```python
df = pd.read_csv("fake_reg_hidden_data.csv")
df = df.sample(frac = 1)
df.reset_index(inplace=True)
        
try:
    df=df.drop(['index'],axis=1)
except:
    pass

df.head()
```

```python
df.isna().sum()
```

```python
df = p_p.pre_process(df, 'Fee(PKR)')
```

```python
df.head()
```

```python

import pickle as pkl
RS = pkl.load(open("scaler_reg.pkl", "rb"))
```

```python
X_ = df.drop('Fee(PKR)', axis=1).values
X_ = RS.transform(X_)
y_ = df['Fee(PKR)'].values
```

```python
model = pkl.load(open("ET_reg.pkl", "rb"))

preds = model.predict(X_)

mse = mean_squared_error(y_, preds)
r2 = r2_score(y_, preds)

print(sqrt(mse))
print(r2)
```
