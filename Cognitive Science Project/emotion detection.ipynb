```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
import nltk
RANDOM_STATE = 42
```

```python
data = pd.read_csv('emotion_dataset.csv')
```

```python
data.head()
```

```python
data['Emotion'].unique()
```

```python
def dummies(x,df):
    temp = pd.get_dummies(df[x], drop_first = True)
    df = pd.concat([df, temp], axis = 1)
    df.drop([x], axis = 1, inplace = True)
    return df
```

```python
data = dummies('Emotion', data)
```

```python
data.info()
```

```python
data.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(data['Text'], data.drop('Text', axis=1), train_size=0.8, random_state=RANDOM_STATE)
```
